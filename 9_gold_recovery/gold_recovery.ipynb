Наша задача подготовить прототип модели машинного обучения, которая должна предсказать коэффициент восстановления золота из золотосодержащей руды. В наших руках данные с параметрами добычи и очистки.

#### Визуализируем концентрацию пропусков

```python
def empty_percent (data):
    empty_percent = pd.DataFrame(data.isna().mean())
    empty_percent.columns = ['empty_percent']
    empty_percent = empty_percent.sort_values('empty_percent', ascending=False)
    return empty_percent[empty_percent['empty_percent'] != 0]
```

```python
train_empty_percent = empty_percent(train)
train_empty_percent.style.background_gradient('coolwarm')
```

```python
test_empty_percent = empty_percent(test)
test_empty_percent.style.background_gradient('coolwarm')
```

Процент пропусков незначительный в обоих выборках, в тесте он не превышает 1%. Наибольшая доля пропусков содержится в трейне и составляет 3%, однако поле "secondary_cleaner.output.tail_sol" содержит 11% пропусков. Для заполнения значений используем KNNImputer. Для пропусков в размере менее 1% используем dropna

```python
for i in range(train_empty_percent.shape[0]):
    if train_empty_percent['empty_percent'].to_list()[i] < 0.01:
        train = train.dropna(subset=[train_empty_percent.index[i]])   
```

```python
train = train.reset_index(drop=True)
```

```python
train_empty_percent = empty_percent(train)
train_empty_percent.style.background_gradient('coolwarm')
```

#### Визуализируем работу KNNImputer

Устраним более значимые пропуски с помощью KNNImputer

```python
train_empty_percent.index
```

```python
imputer = KNNImputer(weights='distance')

ct = make_column_transformer(
    (imputer, ['secondary_cleaner.output.tail_sol', 
               'rougher.state.floatbank10_e_air',
               'rougher.input.floatbank11_xanthate',
               'secondary_cleaner.state.floatbank2_a_air',
               'final.output.concentrate_sol',
               'primary_cleaner.output.concentrate_sol']),
                remainder='passthrough')
```

```python
fil = ct.fit_transform(train)
```

```python
fil = pd.DataFrame(fil).loc[:, 0:5]
fil.columns = train_empty_percent.index
```

```python
fil.head()
```

```python
for i in fil.columns:
    train = train.drop(columns=i)
    train[i] = fil[i]
```

### Напишем функцию вычисления sMAPE

```python
def smape (target, predict):
    return np.sum((np.abs(target - predict) / ((np.abs(target) + np.abs(predict)) / 2))) / len(target) * 100
```

Настроим оценщик для наших моделей на основе функции

```python
smape_scorer = make_scorer(smape, greater_is_better=False)
```

Для начала разделим признаки на фичи и таргет.

Глобально, процесс флотации разделен на два больших блока:
- непосредственно флотация
- процесс последующей очистки

Соответствено, будут построены две модели, которые будут отвечать за первый и второй этап.
